```python
import concurrent
from typing import Iterable

def optimized_multithreading(l: Iterable[int]) -> int:
    with concurrent.futures.ThreadPoolExecutor(max_workers=8) as executor:
        result = sum(executor.map(sum, map(lambda _: l, range(4))))
        return result
```

```python
import warnings
import time
import numpy as np
import seaborn as sns

sns.set_theme(rc={"figure.dpi": 300})

# annoying
warnings.simplefilter(action="ignore", category=FutureWarning)
```

```python
x_simple = [n for n in range(10 ** 6)]

t = time.time()
print(f"{x_simple=}, {optimized_multithreading(x_simple)=}, cost: {time.time() - t}s")
sns.displot(x_simple, kde=True)
```

```python
exponential_sample = np.random.exponential(2, 10 ** 6)

t = time.time()
print(f"{exponential_sample=}, {optimized_multithreading(exponential_sample)=}, cost: {time.time() - t}s")
sns.displot(exponential_sample, kde=True)
```

```python
normal_sample = np.random.normal(size=10 ** 6)

t = time.time()
print(f"{normal_sample=}, {optimized_multithreading(normal_sample)=}, cost: {time.time() - t}s")
sns.displot(normal_sample, kde=True)
```
