# 條件控制式

## `break`

```python
for i in range(1, 6):
    if i == 4:
        break
    print(i)
```

## `continue`

```python
# 避開 i == 4 的條件。

for i in range(1, 6):
    print(i)
    # 為何 i == 4 的時候還輸出呢？
    if i == 4: continue
```

```python
# 避開 i == 4 的條件。

for i in range(1, 6):
    # 如果想要避開 i == 4 的條件，
    # 記得要在 continue 之後才執行 print(i)。
    if i == 4: continue
    print(i)
```

```python
i = 0

while i < 5:
    i += 1
    if i == 4:
        continue
    print(i)
```

```python
i = 0

while i < 5:
    print(i)
    if i == 4:
        continue
    # 因為 i 在後面，所以 i == 4 的時候沒機會繼續 +1
    #   → 無窮迴圈
    i += 1
```
